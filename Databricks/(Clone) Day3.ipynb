```python
from pyspark.sql import SparkSession, functions as F
from pyspark.sql.window import Window

# Initialize Spark
spark = SparkSession.builder.appName("EcommerceAnalysis").getOrCreate()

# Load dataset (adjust path as needed)
events = spark.read.json("ecommerce.json")
products = spark.read.json("products.json")
users = spark.read.json("users.json")
```

```python
# Join events with products and users
events_products = events.join(products, on="product_id", how="inner")
full_data = events_products.join(users, on="user_id", how="left")
```
